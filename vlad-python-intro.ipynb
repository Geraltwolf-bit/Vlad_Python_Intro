```python
print("Hello! Let's get to know each other!")
while True:
    name = input('Enter your name: ').strip()
    if name.replace(' ', '').isalpha() and len(name) > 0:
        break
    else:
        print('Please, enter your name using letters: ')
while True:
    try:
        age = int(input('Enter your age: '))
        break
    except ValueError:
        print('Please, enter your age using digits: ')

while True:
    city = input('Enter your city: ').strip()
    if city.replace(' ', '').isalpha() and len(city) > 0:
        break
    else:
        print('Please, enter your city using letters')

while True:
    color = input('Enter your favorite color: ').strip()
    if color.replace(' ', '').isalpha() and len(color) > 0:
        break
    else:
        print('Please, enter your favorite color using letters')

while True:
    movie = input('Enter your favorite movie: ').strip()
    if movie.replace(' ', '').isalpha() and len(movie) > 0:
        break
    else:
        print('Please, enter your favorite movie using letters')

def age_bucket(age):
    if age < 19:
        return str("You're a teenager!")
    if age < 25:
        return str("You're in your prime!")
    if age < 50:
        return str("You're in a cool age!")
    else:
        return str("Your experince is invaluable")

pitch = age_bucket(age)

print(f"Hello, {name} from {city}! {pitch}")
print(f"You sure you favorite color is {color} and favorite movie is {movie}?")
```

```python
feedback = input("Share your opinion about that movie: ")

def num_words(w):
    return len(w.split())

def num_sentences(s):
    count = 0
    for char in s:
        if char == '!' or char == '?' or char == '.':
            count += 1
    return count
    
def total_length(t):
    return len(t)

def longest_word(lw):
    lw_clean = lw.replace('!', '').replace('?', '').replace('.', '')
    li = lw_clean.split()
    sli = sorted(li, key = len)
    return sli[-1]

w = num_words(feedback)
s = num_sentences(feedback)
l = total_length(feedback)
lw = longest_word(feedback)

print(f"Words: {w} \nSentences: {s} \nTotal length: {l} \nThe longest word: {lw}")
```
